# 03 — Modeling
Train a Logistic Regression baseline and CatBoost advanced model, then compare multiclass metrics.

```python
import sys, pandas as pd
sys.path.append('../src')
from preprocess import preprocess_dataframe
from feature_engineering import engineer_features
from train_model import split_data, train_baseline, train_catboost
frame = engineer_features(preprocess_dataframe(pd.read_csv('../data/sample_data/healthcare_sample.csv')))
X_train, X_test, y_train, y_test = split_data(frame)
baseline = train_baseline(X_train, y_train)
advanced = train_catboost(X_train, y_train)
print('baseline accuracy', baseline.score(X_test, y_test))
print('catboost accuracy', advanced.score(X_test, y_test))
```
